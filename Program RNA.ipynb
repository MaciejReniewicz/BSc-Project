```python
import numpy as np
import pandas as pd
```

```python
dfN = pd.read_csv("set.RNA.all.normal.csv",sep=';', header=0, error_bad_lines=False, low_memory=False, decimal=',')
dfN = dfN.rename(columns = {'Unnamed: 0':'PatientID'})
dfN
```

```python
dfT = pd.read_csv("set.RNA.all.tumor.csv",sep=';', header=0, error_bad_lines=False,low_memory=False, decimal=',')
dfT = dfT.rename(columns = {'Unnamed: 0':'PatientID'})
```

```python
for i in dfT:
    if i!="PatientID" and i!="Tumor":
        dfT[i] = dfT[i].astype('float')
```

```python
Suma = (dfT == 0).sum(axis=0)
count_row = dfT.shape[0]
warunekDFT = dfT.loc[:,Suma>(count_row)/10]
```

```python
dfT.drop(warunekDFT.columns,axis=1, inplace=True)
#dfT
```

```python
for i in dfT:
    if i!="PatientID":
        dfT[[i]] = dfT[[i]].replace(0, np.NaN)
#dfT.isnull().sum()        
```

```python
dfT.fillna(dfT.mean(), inplace=True)    
dfT.insert(1, 'Tumor', 1)
dfT
```

```python
Suma = (dfN == 0).sum(axis=0)
count_row = dfN.shape[0]
warunekDFN = dfN.loc[:,Suma>(count_row)/10]
#Suma
```

```python
#count_row
```

```python
for i in dfN:
    if i!="PatientID":
        dfN[i] = dfN[i].astype('float')
dfN.drop(warunekDFN.columns,axis=1, inplace=True)
#dfN
```

```python
for i in dfN:
    if i!="PatientID":
        dfN[[i]] = dfN[[i]].replace(0, np.NaN)
#dfN.isnull().sum() 
```

```python
#dfN
```

```python
dfN.fillna(dfN.mean(), inplace=True)
dfN.insert(1, 'Tumor', 0)
dfN
```

```python
dfR = pd.concat([dfT, dfN], join='inner')
```
